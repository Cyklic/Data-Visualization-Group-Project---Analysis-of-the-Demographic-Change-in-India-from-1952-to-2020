# **DEFINITION OF TERMS**

Time = The year of the data record, spanning from historical periods to future projections (e.g., 1950-2100).

TotalPopulation = Total population on July 1st, often considered a mid-year estimate. Calculated by adjusting the January population with birth, death, and migration rates.

TotalPopulationMale = Male population on July 1, usually derived from gender distribution statistics in censuses.

TotalPopulationFemale = Female population on July 1, also estimated from gender distribution in censuses.

PopDensity = Population density, calculated as the total population divided by the area of the location.

PopSexRatio = The ratio of males to females in the population, usually expressed as the number of males per 100 females.

MedianAgePop = Median age of the population, calculated as the age that divides the population into two equal halves.

NatChange = Natural population change (births minus deaths).

NatChangeRT = Rate of natural change, typically births minus deaths per 1,000 people.

PopChange = Total population change, including migration.

PopGrowthRate = Annual percentage growth rate of the population, calculated as the percentage increase or decrease in population from one year to the next.

DoublingTime = The estimated number of years it would take for the population to double, assuming the current growth rate continues. It's usually calculated as 70 divided by the growth rate.

Births = Total number of live births in a given year, often based on health and civil registry data.

Births1519 = Number of births to mothers aged 15-19, collected from birth registries.

CrudeBirthRate = Crude birth rate, calculated as births per 1,000 people in a year.

TotalFertilityRate = Total fertility rate, estimating the average number of children a woman would have over her lifetime, based on current age-specific fertility rates.

NetReproductionRate = Net reproduction rate, indicating the average number of daughters a woman would have in her lifetime if she lived to the end of her reproductive years, assuming no mortality.

MeanAgeChildbearing = Mean age at childbearing, calculated as the average age of mothers at the time of birth.

SexRatioatBirth = Sex ratio at birth, typically around 105 male births per 100 female births.

Deaths = Total number of deaths in a year, gathered from civil registries and health agencies.

DeathsMale = Number of male deaths in a year.

DeathsFemale = Number of female deaths in a year.

CrudeDeathRate = Crude death rate, calculated as deaths per 1,000 people per year.

LifeExpectancyatBirth = Life expectancy at birth for the entire population.

LifeExpectancyMale = Life expectancy at birth for males.

LifeExpectancyFemale = Life expectancy at birth for females.

LifeExpectancy15 = Life expectancy at age 15, a measure of the average additional years a 15-year-old is expected to live.

LifeExpectancy15Male = Life expectancy at age 15 for males.

LifeExpectancy15Female = Life expectancy at age 15 for females.

LifeExpectancy65 = Life expectancy at age 65.

LifeExpectancy65Male = Life expectancy at age 65 for males.

LifeExpectancy65Female = Life expectancy at age 65 for females.

LifeExpectancy80 = Life expectancy at age 80.

LifeExpectancy80Male = Life expectancy at age 80 for males.

LifeExpectancy80Female = Life expectancy at age 80 for females.

InfantMortality = Total deaths of infants under age 1.

InfantMortalityRate = Infant mortality rate, calculated as infant deaths per 1,000 live births.

LiveBirthsSurvivingAge1 = Number of live births that survive to age 1.

MortalityUnder5 = Total deaths of children under age 5.

Under5MortalityRate = Probability of dying before age 5, typically per 1,000 live births.

MortalityBefore40 = Probability of dying between ages 0 and 40.

MortalityBefore40Male = Probability of dying between ages 0 and 40 for males.

MortalityBefore40Female = Probability of dying between ages 0 and 40 for females.

MortalityBefore60 = Probability of dying between ages 0 and 60.

MortalityBefore60Male = Probability of dying between ages 0 and 60 for males.

MortalityBefore60Female = Probability of dying between ages 0 and 60 for females.

MortalityBetween15&50 = Probability of dying between ages 15 and 50.

MortalityBetween15&50Male = Probability of dying between ages 15 and 50 for males.

MortalityBetween15&50Female = Probability of dying between ages 15 and 50 for females.

MortalityBetween15&60 = Probability of dying between ages 15 and 60.

MortalityBetween15&60Male = Probability of dying between ages 15 and 60 for males.

MortalityBetween15&60Female = Probability of dying between ages 15 and 60 for females.

NetMigrations = Net migration count, calculated as the difference between immigrants and emigrants.

CrudeNetMigrationsRate = Crude net migration rate, calculated as net migration per 1,000 people.

```python
def is_divisible_by_4(x):
    return x % 4 == 0

finaldf = new_df.loc[(new_df['Time'].apply(is_divisible_by_4)) &
 (new_df['Time'] <= 2020)].sort_values(by='Time').reset_index().drop('index', axis=1)
```
